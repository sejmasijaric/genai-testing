```python
from langtest import Harness
```

```python
harness = Harness(
    task = "question-answering",
    model = {"model": "google-bert/bert-large-uncased-whole-word-masking-finetuned-squad", "hub": "huggingface"},
    data = {"data_source" :"ASDiv","split":"test-tiny"}
)
```

```python
harness.configure(
    {
        'tests': {
            'defaults': {
                'min_pass_rate': 0.65,
                'min_score': 0.8
            },
            'robustness': {
                'uppercase': { 'min_pass_rate': 0.60 },
                'add_abbreviation': {'min_pass_rate': 0.7},
                'add_typo': {'min_pass_rate': 0.7},
                'dyslexia_word_swap': {'min_pass_rate': 0.7}
            }
        }
    }
)
```

```python
harness.data=harness.data[:10]
```

```python
harness.generate()
```

```python
harness.testcases()
```

```python
harness.run()
```
